```python
INDICATORS = [
    "macd", "boll_ub", "boll_lb", "rsi_30",
    "cci_30", "dx_30", "close_30_sma", "close_60_sma"
]
```

```python
fe = FeatureEngineer(
    use_technical_indicator=True,
    tech_indicator_list=INDICATORS,
    use_turbulence=True,          # Thêm turbulence index để đo rủi ro thị trường
    user_defined_feature=False
)
```

```python
df_processed = fe.preprocess_data(df_raw)
```

```python
# Train/Test split
TRAIN_START, TRAIN_END = "2021-01-01", "2025-12-31"
TEST_START, TEST_END   = "2026-01-01", "2026-06-08"
```

```python
df_processed['date'] = pd.to_datetime(df_processed['date'])
```

```python
df_train = data_split(df_processed, TRAIN_START, TRAIN_END)
df_test  = data_split(df_processed, TEST_START, TEST_END)
```

```python
from finrl.meta.env_stock_trading.env_stocktrading import StockTradingEnv
from finrl.agents.stablebaselines3.models import DRLAgent

stock_dim = len(tickers)  # = 10

env_kwargs = {
    "hmax": 100,
    "initial_amount": 1_000_000_000,
    "buy_cost_pct":  [0.0015] * stock_dim,   # ← fix: list 10 phần tử
    "sell_cost_pct": [0.0015] * stock_dim,   # ← fix: list 10 phần tử
    "reward_scaling": 1e-4,
    "state_space": 101,
    "stock_dim": stock_dim,
    "tech_indicator_list": INDICATORS,
    "action_space": stock_dim,
    "num_stock_shares": [0] * stock_dim,
    "print_verbosity": 1,
}

# ============================================================
# FIX 1: Tăng timesteps + điều chỉnh hyperparams
# ============================================================
PPO_PARAMS = {
    "n_steps": 512,          # ← giảm xuống (2048 quá lớn với 10 mã)
    "ent_coef": 0.05,        # ← tăng lên để khuyến khích exploration
    "learning_rate": 1e-4,   
    "batch_size": 64,        # ← giảm xuống
    "gamma": 0.99,
    "gae_lambda": 0.95,
}

# ============================================================
# FIX 4: Retrain với timesteps đủ lớn
# Rule of thumb: stock_dim × 50,000 timesteps tối thiểu
# 10 mã × 50,000 = 500,000
# ============================================================

# Sanity check trước khi train
e_train_gym = StockTradingEnv(df=df_train, **env_kwargs)
obs, _ = e_train_gym.reset()
action = e_train_gym.action_space.sample()
obs2, reward, done, trunc, info = e_train_gym.step(action)
print("✓ Environment OK — reward:", round(reward, 4))

# Train
env_train, _ = e_train_gym.get_sb_env()
agent = DRLAgent(env=env_train)
model_ppo = agent.get_model("ppo", model_kwargs=PPO_PARAMS)
trained_ppo = agent.train_model(model=model_ppo, tb_log_name="ppo_vn", total_timesteps=500_000)
```
